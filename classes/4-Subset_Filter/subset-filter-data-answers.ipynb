```python
# practical - load wine dataset
wine = pd.read_csv('datasets/winemag-data-130k-v2.csv', 
        usecols=col[1:])
display(wine.head(3))
display(wine.columns)
```

```python
wine_subset = wine[['country', 'description', 'points', 'price', 'title']]
wine_subset.head()
print(len(wine_subset))
```

```python
# filter US only
wine_us = wine_subset[wine_subset['country']=='US']
print(len(wine_us))
```

```python
# filter US only and points >= 90
wine_us = wine_subset[(wine_subset['country']=='US')&(wine_subset['points']>=90)]
print(len(wine_us))
```

```python
wine['country'].unique()
```

```python
# filter Germany or France 
wine_eu = wine_subset[(wine_subset['country']=='France')|(wine_subset['country']=='Germany')]
print(len(wine_eu))
```

```python
# filter Germany or France or Brazil
countries = ['France', 'Germany', 'Brazil']

wine_sub = wine_subset[wine_subset['country'].isin(countries)]
print(len(wine_sub))
```

```python
wine_sub['country'].unique()
```

```python
wine_sub['title'].unique()
```

```python
# wine 2012 - title contains 2012 
wine2012 = wine_sub[wine_sub['title'].str.contains('2012')]
wine2012.head()
```
