#### 5. Break and Continue

```python
dims = [[5, 10], [5, 8], ['10', 'NA'], [7, 9]]
```

```python
# break
for element in dims:
    if type(element[0]) == int:
        area = get_reg_area(element[0], element[1])
        print(area)
    else:
        break
```

```python
# continue
for element in dims:
    if type(element[0]) == int:
        area = get_reg_area(element[0], element[1])
        print(area)
    else:
        continue
```

#### 6. Putting everything together

- input: height, width (list of values)
- output: area, define the type of a room

```python
def define_room(dims):
    for element in dims:
        if type(element[0]) == int:
            area = get_reg_area(element[0], element[1])
            if area >= 45:
                print(area, 'Deluxe')
            else:
                print(area, 'Standard')
        else:
            continue
```

```python
define_room(dims)
```
